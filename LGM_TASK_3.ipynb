```python
# Visualizing data for further understanding

sns.countplot(x='region_txt',data=Terrorism_data)
plt.xticks(rotation=90)
plt.xlabel('Region of Attacked')
plt.ylabel('Most Attacked Region')
plt.show()
```

From above graph we can conclude that Western Europe was attacked more times bt terrorist

```python
sns.countplot(x='iyear',data=Terrorism_data)
plt.xticks(rotation=90)
plt.xlabel('Year')
plt.ylabel('No Of Attack')
plt.title('Terrorist Attack Per Year')
plt.show()
```

From above graph we can conclude that in the year 1978 more no of attacks occur

```python
sns.countplot(x='attacktype1_txt',data=Terrorism_data,palette='PuBuGn_r')
plt.xticks(rotation=90)
plt.xlabel('Attack Type')
plt.ylabel('No Of Attack')
plt.title('Type of weapon use in Attack')
plt.show()
```

From above graph we can conclude that most of attacks happen by Explosion/Bombing

```python
sns.countplot(x='targtype1_txt',data=Terrorism_data,palette='Spectral')
plt.xticks(rotation=90)
plt.xlabel('Type')
plt.ylabel('Count')
plt.title('Target of Terrorist')
plt.show()
```

From above graph we can conclude that Business area is attacked most by terrorist

```python
plt.figure(figsize=(12,10))
data['success'].value_counts().plot.pie(autopct='%.2f')
plt.show()
```
